```python
import pandas as pd
import numpy as np
```

```python
#mpg.csv dosyasını okuyun ve df değişkenine atayın.
```

```python
df = pd.read_csv("mpg.csv")
df.head()
```

```python
#1) df'in içindeki sütunların birbirleriyle olan korelasyonlarına(ilişkilerine) bakın.
df.corr()
```

```python
#2) mpg sütunu ile diğer sütunların arasındaki korelasyonları özel olarak inceleyin.
#(Bunu göstermedik ama df.corr() çağırdıktan sonra aynı normal df'lerde olduğu gibi istediğiniz sütunu çağırabiliyorsunuz.)
df.corr(numeric_only= True)["mpg"]
```

```python
#3) cylinders sütununda hangi unique değerler olduğuna liste halinde bakın.
df["cylinders"].unique()
```

```python
#4) cylinders sütunundaki unique değerlerin her birinden kaçar adet olduğuna bakın.
df["cylinders"].value_counts()
```

```python
#5) Arabanın modeli 75-80 yılları arasındaki arabaların hepsini getirin.(75 ve 80 dahil)
df[df["model_year"].between(75, 80)]
```

```python
#6) horsepower'ı 155-180 arasındaki arabaların hepsini getirin.
#(155 dahil, 180 dahil değil)
#(Bu sorguyu yapabilmeniz için horsepower sütununu object durumdan float duruma getirmeniz gerekiyor.)
df["horsepower"].replace("pd.na", np.nan, inplace= True)
df["horsepower"] = df["horsepower"].astype(float)
print(df["horsepower"].dtype)
df[df["horsepower"].between(155, 180, inclusive="left")]
```

```python
#7) Elinizdeki df'i horsepower'a göre büyükten küçüğe sıralayın ardından df'in indexlerini sıfırlayın.
df.sort_values(by= "horsepower", ascending= False).reset_index()
```

```python
#8) Elinizdeki df'den en küçük 5 acceleration'a sahip olan arabaları getirin.
df.nsmallest(5, columns= "acceleration")
```

```python
#--- CIA_Country_Facts.csv dosyasını okuyun ve yeniden df değişkenine atayın.(
#Artık eski df ile işimiz kalmadığı için aynı ismi verebiliriz.)

df = pd.read_csv("CIA_Country_Facts.csv")
df.head()
```

```python
#9) Elimizdeki CIA'in ülkeler hakkında topladığı basit açık kaynak verileri df'inde kaç ülkenin verisi olduğunu bulun.
df["Country"].info()
```

```python
#10) Elimizdeki sütunlardan hangilerinde kaçar tane nan veri olduğunu bulun.
df.isna().sum()
```

```python
#11) Elimizde tamamen nan değerlerden oluşan bir satır varsa o satırı düşürün.
#(Tamamen nan değerlerden oluşan bir satır olup olmadığını, herhangi bir satırın düşüp düşmediğini kontrol edin)
print(df.info())
df.dropna(how= "all", inplace= True)
print("#####################################")
print(df.info())
```

```python
#12) Her Bölgedeki(Region) toplam kişi sayısına bakın.
df.groupby("Region")["Population"].sum()
```

```python
#13) Climate değerleri nan olanlar, tüm df'in ortalama climate değeriyle doldurun.
#(Gerçek bir analiz yapıyor olsaydık bu çok mantıklı bir hareket olmazdı,
#belki her bulunduğu bölgenin ortalama climate değeriyle doldurmak daha mantıklı bir hareket olurdu. 
#Fakat şimdilik bu kadar kasmaya gerek yok.
#nan değerleri tüm climate ortalamalarıyla doldursanız yeterli.)

df.Climate.fillna(df["Climate"].mean(), inplace= True)
df["Climate"].isna().value_counts()
```

```python
#14) Bölgeler ve Ülkelere göre gruplama yapın ardından Doğum miktarlarına bakın.
birthrate_by_region_country = df.groupby(["Region", "Country"])["Birthrate"].sum()
birthrate_by_region_country.sort_values(ascending = False)
```

```python
#15) Veriyi istediğiniz şekilde inceleyerek bu zamana kadar öğrendiğiniz bilgilerle anlamlı 5 adet yargı cümlesi çıkartın.
#(1) En çok insan Asya kıtasında yaşıyor,
#2) En çok insan Afrinın sahralarında ölüyor,
#3) Ortalama en yüksek telefon sahibi Kuzey Amerikada,
#4) En çok okuma yazma oranına sahip ülkeler Baltık ülkeleri vb.)
# 1) En çok insan Asya kıtasında yaşıyor
sum_of_Population_by_region = df.groupby("Region")["Population"].sum()
sum_of_Population_by_region.sort_values(ascending= False)
```

```python
#2) En çok insan Afrinın sahralarında ölüyor
df.sort_values(by = "Deathrate", ascending= False).reset_index()
```

```python
#3) Ortalama en yüksek telefon sahibi Kuzey Amerikada
phones_mean = df.groupby("Region")["Phones (per 1000)"].mean()
phones_mean.sort_values(ascending = False)
```
